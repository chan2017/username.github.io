First, we have to ready our environment for our <i>data science experiment</i>. Importing modules make projects much easier so that we can easily use functions that have already been written by other people to facilitate tasks. <p>

One of the modules we have imported is Requests. This is used to handle HTTP requests from Python. It allows our script to have browser-like capabilities in getting or posting to websites. ```python

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
import numpy as np
import zipfile
import os
import shutil

import matplotlib.pyplot as plt
import seaborn as sns

```

The first step in scraping data is getting the information from our target web page. We will use the aforementioned REQUESTS module to obtain the HTML used to generate the web page. <br><br>
The EPA website has a collection of downloaded fuel economy conveniently organized in a table. To our human eyes, this table looks easy to navigate and to understand. However, a computer sees it very differently. Content in HTML tables are listed by rows, which is contrary to the column oriented way it is presented. <br><br>
This is where the aforementioned module Beautiful Soup comes in handy. Beautiful Soup reads in our requested page's content from REQUESTS and saves it in our scripting environment. We have attached an option to our Beautiful Soup object, letting it know that we are reading in HTML. Using "lxml" as an option is preferred over using HTML.parser due to its speed in reading the request into our Beautiful Soup object.

```python
url = 'http://www.fueleconomy.gov/feg/download.shtml'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'lxml')
```

```python
soup
```

While humans can easily parse information from a variety of formats, computers find this job much more difficult. Luckily, our target data files all have the same string format. Not only that, but other data files have markedly different formats. This allows us to use RE to make regular expressions (regex) to glean our source URLs from the web page's HTML.<br><br>

```python
reg = 'epadata/\d{2}data.zip' 

links = soup.find_all('a')
zlist = []
for tag in links:
    link = tag.get('href',None)
    if link is not None:
        if re.match(reg, link):
            zlist.append(link)
```

Our target files are stored elsewhere on the Internet. From navigating the original EPA site, we can find that their links point to a subdirectory of another government website "www.fueleconomy.gov". We attach the address of this subdirectory to the names of the target files we collected from the last step. This will give us the entire address of the target files themselves. <br><br>
In the browser, files are immediately opened when open their addresses. However, we are using a Python script. We need to use REQUESTS again to prompt the website to give us a data stream to our target file. Due to the large size of these files, we will process them in chunks. <br><br>
After we have downloaded our target file, we have to extract the data inside of it. The data we are looking for is zipped with other files and given a nonuniform name. It would be difficult to use this name in the future, so we will extract them individually to folders bearing their zip file name (remember that the zip file was nicely formatted? We will take advantage of this in the next step). 



```python
suf = 'http://www.fueleconomy.gov/feg/'

filelst = []
for x in zlist:
    url = suf+x
    localfn = url.split('/')[-1]
    yearno = int(localfn[:2])
    if yearno < 50 and yearno > 00:
        r = requests.get(url, stream=True)
        with open(localfn, 'wb') as f:
            for chunk in r.iter_content(chunk_size=1024): 
                if chunk: # filter out keep-alive new chunks
                    f.write(chunk)
        zip_ref = zipfile.ZipFile(localfn, 'r')
        lcl = "dldir"+localfn
        filelst.append(lcl)
        zip_ref.extractall(lcl)
        zip_ref.close()
```

Before we actually process our data files, we should create a directory to organize them. This makes our workspace neater. <br><br>
The following script creates a folder within our current working directory if it doesn't already exist. Notice that the name of our directory doesn't matter - since we are working in our local machine, we can start using our own variables.

```python
datdir = "shibe_data"
if not os.path.exists(datdir):
    os.makedirs(datdir)
```

Our data spread covers the early aughts when .xls was the common MS Excel file format. Eventually, .xlsx eclipsed .xls files by 2010. Another popular format of data storage is Comma Separated Values (CSV). CSV takes the appearance of a text file which is able to be used by text editors as well as spreadsheet software. <br><br>
In our sample of data files, all three formats are represented. We will be able to take advantage of this fact to identify our desired data file from the other metadata included in the original zip file. However, we will have to consider their file extensions when renaming them to formatted yet descriptive names. <br><br>
We construct a regex which identifies our selected file extensions which will specifically look at the ending of our string. We search through all the folders that we have extracted zipfiles to and cycle through all the names of the files within. Any name which contains our regex will be captured and renamed with a version of our folder name (this is to ensure that the year the data is from is easily identifiable from the name). With the following script we will also take advantage of the OS module's rename function, which allows us to move files as well as rename them. <br><br>

```python
reg = "(.xlsx|.xls|.csv)$"
for item in filelst:
    for filename in os.listdir(item):
        curname =  item + "\\" + filename 
                
        m = re.search(reg, filename)
        if m:
            newname = datdir + "\\" + item[:-4] + m.group(0)
            os.rename(curname, newname)
```

In the process of obtaining our desired data files, we have created many unnecessary files. This clutters our current working directory. This following script removes all the zip files in our current directory and removes all the folders created through zip file extraction. <br><br> 
We start by getting all the names in our current working directory. Files that are identified as zip files are removed. Directories that bear the prefix used in extracting zip files are also removed. We can be sure that no important information is discarded at this point, as we no longer have use for any zip file, and our data folder does not share the same name as our special prefix.

```python

lst = os.listdir(os.getcwd())
for x in lst:
    m = re.search(".zip$", x)
    if m:
        n = re.search("^dldir",x)
        if n:
            shutil.rmtree(x)
        else:
            os.remove(x)
```

```python
# Reads data files into Pandas Dataframes according to file type (.csv .xlx .xlsx). The df are stored in a temporary dict 
# according to their respective years. The original zip file from the EPA site does not give year in YYYY format, so we have 
# chosen not to use the YYYY format until this point where it becomes necessary and after we have gleaned out only the years 
# prefixed by "20-"d
# Some lines of the original data files are formed incorrectly. Adding the option "error_bad_lines=False" skips these lines
```

```python
%%capture
datfiles = os.listdir(datdir)
pdfiles = {}
for x in datfiles:
    y = re.search("[0-9]{2}",x)
    year = "20" + y.group(0)
    cs = re.search(".csv$", x)
    if cs:
        newp = pd.read_csv(datdir + "\\" + x, error_bad_lines=False)
    else:
        newp = pd.read_excel(datdir + "\\" + x)
        
    pdfiles[year] = newp
    
```

```python
# Went through all the data files from 2001-2018 and found all the persisting variables and listed their aliases in a dict, vars
```

```python
varss = {}
varss['Class'] = ['CLASS','Carline Class Desc', 'Class']
varss['Manufacturer'] = ['MFR', 'Manufacturer', 'Mfr Name', 'Mfr Name ']
varss['carline name'] = ['CAR LINE', 'Carline', 'carline name']
varss['cmb'] = ['COMB MPG (GUIDE)', 'Comb FE (Guide) - Conventional Fuel', 'cmb']
varss['cty'] = ['CITY MPG (GUIDE)', 'City FE (Guide) - Conventional Fuel', 'cty']
varss['cyl'] = ['# Cyl', 'NUMB CYL', 'cyl']
varss['displ'] = ['DISPLACEMENT', 'Eng Displ', 'displ']
varss['fcost'] = ['ANL FL CST', 'Annual Fuel1 Cost - Conventional Fuel', 'fcost']
varss['fl'] = ['FUEL TYPE', 'Fuel Usage Desc - Conventional Fuel', 'fl']
varss['hwy'] = ['HWY MPG (GUIDE)', 'Hwy FE (Guide) - Conventional Fuel', 'hwy']
varss['trans'] = ['TRANS', 'Trans as listed in FE Guide (derived from col AA thru AF)', 'Trans in FE Guide (MFR entered for data entered after May 13 2011)', 'Transmission', 'trans']
varss['ucmb'] = ['Comb Unadj FE - Conventional Fuel', 'UNRND COMP (EPA)', 'ucmb']
varss['ucty'] = ['City Unadj FE - Conventional Fuel', 'UNRND CITY (EPA)', 'ucty']
varss['uhwy'] = ['Hwy Unadj FE - Conventional Fuel', 'UNRND HWY (EPA)', 'uhwy']
varss['drv'] = ['DRIVE SYS', 'Drive Sys', 'drv']
```

```python
# Finds all the column names which are included inside of the shared variable map and replaces the dataframe 
# with only those columns. Renames them to the shared variable map's key names, so that column names can be 
# standardized. The columns are reordered before replacing the dataframes inside the original master map of df
```

```python
varlst = list(varss.keys())
yearlst = list(pdfiles.keys())


for z in yearlst:
    curpd = pdfiles[z]
    cols = list(curpd.columns.values)
    checklst = []
    rename = []
    
    for x in cols:
        for y in varlst:
            if x in varss[y]:
                checklst.append(x)
                rename.append(y)
    curpd = curpd[checklst]
    curpd.columns = rename
    curpd = curpd[varlst]
    curpd['year'] = z
    pdfiles[z] = curpd
```

```python
# Finds all car classes from over the years. Aliases for car classes are 
# combined in the carclass dict. Some types (small SUV and standard SUV are 
# combined since they were originally the same category in previous years).
```

```python
cartypes = []
for x in yearlst:
    curpd = pdfiles[x]
    temp = set(curpd['Class'].values)
    cartypes.extend(list(temp))
cartypes = list(set(cartypes))
```

```python
carclass = {}

carclass['COM'] = ['COMPACT CARS','Compact Cars','SUBCOMPACT CARS','Subcompact Cars','MINICOMPACT CARS','Minicompact Cars']
carclass['LAR'] = ['LARGE CARS','Large Cars']
carclass['MID'] = [ 'MIDSIZE CARS','Midsize Cars']
carclass['MINIV'] = ['MINIVAN - 2WD','Special Purpose Vehicle, minivan 2WD','SPEC PURP VEH - MINIVAN - 2WD','SPEC PURP VEH - MINIVAN - 4WD','Special Purpose Vehicle, minivan 4WD','MINIVAN - 4WD']
carclass['PU'] = ['SMALL PICKUP TRUCKS 4WD','Small Pick-up Trucks 4WD', 'Small Pick-up Trucks 2WD','STANDARD PICKUP TRUCKS 2WD','Standard Pick-up Trucks 2WD','Standard Pick-up Trucks 4WD']

carclass['SW'] = ['MIDSIZE STATION WAGONS','SMALL STATION WAGONS','Small Station Wagons','Midsize Station Wagons']
carclass['SUV'] = ['SPEC PURP VEH - S.U.V. - 4WD','S.U.V. - 4WD','Small SUV 4WD','Special Purpose Vehicle, SUV 4WD','Standard SUV 4WD','SPEC PURP VEH - S.U.V. - 2WD','S.U.V. - 2WD','Small SUV 2WD','Special Purpose Vehicle, SUV 2WD','Standard SUV 2WD']

carclass['2S'] = ['TWO SEATERS','Two Seaters']
```
